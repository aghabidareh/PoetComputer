```python
import sqlite3
import pandas as pd
import numpy as np
```

```python
path = 'ganjoor.s3db'
output_file = "poems.txt"
connection = sqlite3.connect(path)
cursor = connection.cursor()
query = "SELECT * FROM verse;"
cursor.execute(query)
rows = cursor.fetchall()
```

```python
with open(output_file, "w", encoding="utf-8") as file:
    for row in rows:
        verse = row[3].strip()
        file.write(verse + "\n")
```

```python
connection.close()
```

```python
print(f"the all of verses saved in {output_file}.")
```
